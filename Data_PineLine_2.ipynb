```python
for file in os.listdir(path):
    file_path = path + file
    file_path.replace(' ','')
    print(file_path)
```

```python
frames = []
for file in os.listdir(path):
    file_path = path + file
    df = pd.read_excel(file_path)
    frames.append(df)
    result = pd.concat(frames)
result
```

```python
df_1 = result.copy()
df_1.columns
```

```python
new_change = df_1['Change (+ / -%)'].apply(lambda x: int(x.replace(',','').split('(')[0]))
plus_minus = df_1['Change (+ / -%)'].apply(lambda x: float(x.replace(')','').replace('%','').replace(',','').split('(')[1]))
df_1['New_change'] = new_change
df_1['Plus_Minus'] = plus_minus
```

```python
columns = ['Right', 'StockCode', 'Closing price',
       'vol ( GD trading )', 'val ( GD trading )', 'vol ( GD agreement )',
       'vol ( GD agreement )_1', 'Opening price', 'Highest price',
       'Lowest price', 'New_change', 'Plus_Minus']
df1 = df_1[columns]
df1.head()
```

```python
def connect_database():
    drivers = 'ODBC Driver 17 for SQL Server'
    SeverName = 'DESKTOP-L8OJDR9\SQLEXPRESS01'
    db = 'CHUNGKHOAN'
    connection = pyo.connect(driver = drivers, 
                            host = SeverName ,database = db ,trusted_connection = 'yes')
    conn = connection
    cur = conn.cursor()
    return cur,conn
```

```python
cur , conn = connect_database()
```

```python
def create_table(cur,conn,query):
    cur.execute(query)
    conn.commit()
def insert_table(cur,conn,query,df1):
    for i,row in df1.iterrows():
        cur.execute(query,list(row))
    conn.commit()
```

```python
new_month = month.replace('-','_')
query_create_table = (f""" CREATE TABLE Information_{new_month} (
                            Day_Trading datetime,
                            StockCode varchar(20),
                            Closing_Price int,
                            GD_trade_vol int,
                            GD_trade_val int,
                            GD_agreement_vol int,
                            GD_agreement_vol_1 float,
                            Opening_Price int,
                            Highest_Price int,
                            Lowest_Price int,
                            New_Change int,
                            Plus_Minus varchar(20)
                        )""")
query_insert_table = (f""" INSERT INTO Information_{new_month} (
                            Day_Trading ,
                            StockCode ,
                            Closing_Price ,
                            GD_trade_vol ,
                            GD_trade_val ,
                            GD_agreement_vol ,
                            GD_agreement_vol_1 ,
                            Opening_Price ,
                            Highest_Price ,
                            Lowest_Price ,
                            New_Change ,
                            Plus_Minus )
                        VALUES (?,?,?,?,?,?,?,?,?,?,?,?)""")
```

```python
create_table(cur,conn,query_create_table)
```

```python
insert_table(cur,conn,query_insert_table,df1)
```

```python
cur.close()
conn.close()
```
